```python
#下載中文字型
import wget
wget.download("https://github.com/roberthsu2003/machine_learning/raw/refs/heads/main/source_data/ChineseFont.ttf")
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mpl

fontManager.addfont("ChineseFont.ttf")
mpl.rc ('font', family='ChineseFont')

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, marker='o', label='測試', linestyle='-.', color='r', markersize=8, linewidth=3)
axes.set_title("Line Plot on Existing Axes")
axes.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mpl

fontManager.addfont("ChineseFont.ttf")
mpl.rc ('font', family='ChineseFont')

figure, axes = plt.subplots()
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x, y, marker='o', label='學生1', linestyle='-.', color='r', markersize=8, linewidth=3)
axes.set_title("折線圖")
axes.set_xticks([0,1,2], ["一","二","三"])
axes.set_yticks([3,4,5], ['三','四','五'])
axes.legend()
plt.show()
```
